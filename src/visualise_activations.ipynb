```python
import numpy as np
import matplotlib.pyplot as plt 
from layers.activation.sigmoid_layer import *
from layers.activation.relu_layer import *
from layers.activation.tanh_layer import *
from layers.activation.leaky_relu_layer import *
%matplotlib inline

x = np.arange(-12.0, 12.0, 0.2)
```

```python
sigmoid_layer = Sigmoid_layer()

plt.plot(x, sigmoid_layer.forward(x))
plt.title("Sigmoid forward")
plt.show()
```

```python
plt.plot(x, sigmoid_layer.backward(x))
plt.title("Sigmoid backward")
plt.show()
```

```python
tanh_layer = Tanh_layer()

plt.plot(x, tanh_layer.forward(x))
plt.title("Tanh forward")
plt.show()
```

```python
plt.plot(x, tanh_layer.backward(x))
plt.title("Tanh backward")
plt.show()
```

```python
relu_layer = Relu_layer()

plt.plot(x, relu_layer.forward(x))
plt.title("Relu forward")
plt.show()
```

```python
plt.plot(x, relu_layer.backward(x))
plt.title("Relu backward")
plt.show()
```

```python
leaky_relu_layer = Leaky_relu_layer(alpha=0.1)

x = np.arange(-12.0, 12.0, 0.2)
plt.plot(x, leaky_relu_layer.forward(x))
plt.title("Leaky ReLU forward")
plt.show()
```

```python
plt.plot(x, leaky_relu_layer.backward(x))
plt.title("Leaky ReLU backward")
plt.show()
```
